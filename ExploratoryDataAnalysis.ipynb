```python
fig,axes=plt.subplots(1,1,figsize=(12,7))
correlation=df.corr()
sns.heatmap(correlation)
plt.show()
```

```python
fig,axis=plt.subplots(1,2,figsize=(18,5))
sns.barplot(data=df.groupby('Sub-Category')['Sales','Profit'].agg(sum),x='Sales',y='Profit',ax=axis[1])
df.groupby('Sub-Category')['Sales','Profit'].agg(sum).plot(kind='bar',ax=axis[0])
plt.xticks(rotation=90)
plt.show()

```

```python
fig,axes=plt.subplots(1,2,figsize=(15,5))
sns.barplot(data=df.groupby('Sub-Category')['Quantity','Sales'].agg(sum),x='Sales',y='Quantity',ax=axes[1])
df.groupby('Sub-Category')['Quantity','Sales'].agg(sum).plot(kind='bar',ax=axes[0])
plt.xticks()
plt.show()
```

--> Very low quantity has high sales.

```python
fig,axes=plt.subplots(1,2,figsize=(15,5))
df.groupby('Sub-Category')['Profit','Quantity'].agg(sum).plot(kind='bar',ax=axes[0]).set_title('Quantity and Profit relation based on sub category')
df.groupby('Sub-Category')['Profit','Discount'].agg(sum).plot(kind='bar',ax=axes[1]).set_title('Profit and Discount based on sub  category')
plt.xticks()
plt.show()
```

```python
fig,axes=plt.subplots(2,2,figsize=(16,8))
fig.suptitle("Distribution Plots")
sns.distplot(df['Sales'],ax=axes[0,0])
sns.distplot(df['Profit'],ax=axes[0,1])
sns.distplot(df['Discount'],ax=axes[1,0])
sns.distplot(df['Quantity'],ax=axes[1,1])
plt.show()
```

--> Sales is rightly skewed. Profit is normally distributed. Discount density is highest at 0.0 and 0.2. Quantity density is highest  in range(2-4)

```python
fig,axes=plt.subplots(2,2,figsize=(14,8))
fig.suptitle("Profit vs Shipping mode")
sns.barplot(df['Ship Mode'],df['Profit'],ax=axes[0,0])
sns.lineplot(df['Ship Mode'],df['Profit'],ax=axes[0,1])
sns.barplot(df['Segment'], df['Profit'],ax=axes[1,0])
sns.lineplot(df['Segment'],df['Profit'],ax=axes[1,1])
plt.show()
```

--> In first class shipping mode the profit is highest .
 Consumer segment has least profit and Home office has highest.

```python
fig,axis=plt.subplots(1,1,figsize=(10,5))
sns.lineplot(df['Discount'],df['Profit'])
```

--> when discount <=0.2, sales was in profit but when discount>0.2, sales faces huge loss.
